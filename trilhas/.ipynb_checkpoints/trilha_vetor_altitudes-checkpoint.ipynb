```python
trilha = open("cunha-paraty-caminho-do-ouro.tcx")
altitudes = []
for line in trilha.readlines():
    l = line.split()
    n = ''
    if l[0][:16] == '<AltitudeMeters>':
        i = 0
        while l[0][16+i] != '<':
            n += l[0][16+i]
            i = i+1
        altitudes.append(n)

altitudes
```
